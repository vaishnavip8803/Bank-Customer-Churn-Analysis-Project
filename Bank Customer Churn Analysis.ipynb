# ***Bank Customer Churn Analysis Project***

# **Problem Statement**

Many banks lose customers because they switch to other banks or stop using banking services. Customer churn reduces the bank's revenue and increases the cost of acquiring new customers. The objective of this project is to analyze customer data, identify the factors that influence customer churn, and provide insights that help the bank improve customer retention.

# **Objectives**
1) Analyze the bank customer dataset to understand customer behavior.
2) Clean and prepare the data by handling missing values, duplicates, and unnecessary columns.
3) Perform Exploratory Data Analysis (EDA) using charts and graphs.
4) Identify the factors that are related to customer churn, such as age, balance, country, gender, and account activity.
5) Generate useful business insights that can help the bank reduce customer churn and improve customer satisfaction.

# 1) CountPlot

```python
sns.countplot(x="Churn",data=data)
plt.title("Churn Count")
plt.xlabel("Churn (0 = No, 1 = Yes)")
plt.ylabel("Number of Customers")
plt.show()
```

# Observations

1) Most customers did not churn (0)
2) Some customers churned (1) 
3) Data is imbalanced
4) Bank should focus on customer retention

# 2) Age Distribution

```python
sns.histplot(data["Age"], bins=10)
plt.title("Age Distribution")
plt.show()
```

# Observations

1) Most customers are in the 30–50 age group.
2) Data is centered around middle age.

# 3) Credit_Score Distribution

```python
data['Credit_Score'].hist()
plt.title("Credit_Score Distribution")
plt.xlabel("Credit_Score")
plt.ylabel("Count")
plt.show()
```

2) Very few customers have very low or very high scores.
3) Data is normally distributed (bell-shaped).

# 4) Account_Balance Distribution

# ***Step11 : Bivariate Analysis***

# 1) Gender Vs Churn

```python
sns.countplot(x="Churn",hue="Gender",data=data)
plt.title("Gender Vs Churn")
plt.show()
```

# Observations

1) Both male and female customers are present in large numbers.
2) female in this dataset shows slightly higher churn.

# 2) Country Vs Churn

```python
sns.countplot(x="Country", hue="Churn", data=data)
plt.title("Country vs Churn")
plt.show()
```

# Observations

1) Churn varies across different countries.
2) One country (usually Germany in this dataset) shows higher churn rate.
3) Other countries have lower or moderate churn.

# 3) Age vs Churn

```python
sns.boxplot(x='Churn', y='Age', data=data)
plt.title("Age vs Churn")
plt.show()
```

# Observations

1) Customers who churned are generally older.
2) Median age of churned customers is higher than non-churned.
3) Younger customers are less likely to churn.

# 4) Account Balance vs Churn

```python
sns.boxplot(x='Churn', y='Account_Balance', data=data)
plt.title("Balance vs Churn")
plt.show()
```

# Observations

1) Customers who churned often have higher balances.
2) Non-churn customers include many with low or zero balance.

# 5) Active Member vs Churn

```python
sns.countplot(x='Is_Active_Member', hue='Churn', data=data)
plt.title("Active Member vs Churn")
plt.show()
```

# Observations

1) Inactive customers (0) have higher churn.
2) Active customers (1) mostly stay with the bank.

# 6) Number of Products vs Churn

```python
sns.countplot(x='Number_of_Products', hue='Churn', data=data)
plt.title("Products vs Churn")
plt.show()
```

# Observations

1) Customers with 1 product show higher churn.
2) Customers with 2 products are more stable (less churn).
3) Very high product counts may also show some churn.

# ***Step12 : Check Outliers***

```python
sns.boxplot(x="Account_Balance",data=data)
plt.title("Account_Balance Outliers")
plt.show()
```

# Observations

1) Presence of outliers (very high balances).
2) Most customers have low to medium balance.
3) Few customers hold extremely high amounts.
